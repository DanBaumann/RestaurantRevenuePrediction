Statistical Testing:

I will use a D'Agostino's  𝐾2  Test to check if the target variable exhibits a Gaussian-like distribution

We can make the following interpretations about p-values in statistical testing:

If  𝑝<𝑎𝑙𝑝ℎ𝑎 : reject H0 that the distribution is normal i.e. it is not normal
If  𝑝>𝑎𝑙𝑝ℎ𝑎 : fail to reject H0 that it is normal, i.e. it is norma

```python
# restaurant appears fairly Gaussian, of which we can do a normality test
from scipy import stats

def normality_test(data):
    statistic, p = stats.normaltest(data)
# using an alpha level of 0.5
    alpha = 0.05
    if p > alpha:
        print("Data is Gaussian: Failure to reject the null hypothesis")
    else:
        print('Data is not Gaussian: reject the null hypothesis, and accept the alternative')
    return statistic, p
```

```python
statistic, p = normality_test(train_revenue)

print("p-value is {}".format(p))
print("Kurtosis of revenue: {}".format(stats.kurtosis(train_revenue)))
print("Skewness of revenue: {}".format(stats.skew(train_revenue)))
```

We can use metrics to show deviation away from normal distributions such as skewness and kurtosis

Kurtosis describes how heavy the tails of distributions are
  * A normal distribution has a kurtosis of close to zero, a value greater than this indicate heavy tails, whereas a value less than zero implies light tails
Skewness measures the asymmetry of the distirbtuion
  * Value between -0.5 and 0.5 show that the distribution is fairly symmetrical
So, all three 'tests' inform us that the distribution of global active power is NOT normal

<h2 align="center">Regularisation Model: Ridge</h2>

In ridge regression, the linear regression cost function is changed by adding a penalty term to the square of the magnitude of the coefficients.
If you have two predictors the full equation would look like this (Notice that there is a penalty term `m` for each predictor in the model - in this case, two) : 

$$ \text{cost_function_ridge}= \sum_{i=1}^n(y_i - \hat{y})^2 = $$

$$ \sum_{i=1}^n(y_i - ((m_1x_{1i})-b)^2 + \lambda m_1^2 + (m_2x_{2i})-b)^2 + \lambda m_2^2)$$

$$\lambda \text{ is a so-called hyperparameter, which needs to be specified when running a model}$$
A small lambda will make a ridge model look like a linear regression model (i.e. small penalisation). For large lambda, coefficients will be more heavily penalised

```python
# instantiating the model and setting up a grid search
ridge_model = Ridge()
ridge_model_grid = [{'normalize':[True, False], 'alpha':np.logspace(0,10)}]
grid_search = GridSearchCV(estimator=ridge_model, param_grid=ridge_model_grid, scoring='neg_mean_squared_error',
                          cv=5)
# fitting the model
grid_search.fit(X, y)
# printing the best parameters
print("Best Parameters found with GridSearchCV:")
print(grid_search.best_params_)
```

```python
# sanity check. Getting the cross_validated results and creating a dataframe
cv_results = grid_search.cv_results_
scores_df = pd.DataFrame(cv_results).sort_values(by='rank_test_score')

scores_df
```

```python
# saving the best parameters in our model
ridge_model.set_params(**grid_search.best_params_)
# and fitting the initial model with the best parameters saved
ridge_model.fit(X, y)

print("Training MSE: {}".format(mean_squared_error(y, ridge_model.predict(X))))
print("Training RMSE: {}".format(np.sqrt(mean_squared_error(y, ridge_model.predict(X)))))


ridge1_predictions_df = pd.DataFrame(columns=['Prediction'],  index=test.index, data=ridge_model.predict(test))
ridge1_predictions_df['Prediction'] = ridge1_predictions_df['Prediction'].apply(np.square)
ridge1_predictions_df.describe().astype(int)
```

```python
# saving predictions as a csv
ridge1_predictions_df = ridge1_predictions_df.reset_index(drop=True)
ridge1_predictions_df.index.name='Id'
ridge1_predictions_df.to_csv('ridge1_predictions.csv')
```

<h2 align="center">Concluding Thoughts</h2>
<br>
The best RMSE score I got with the various models was with the first Ridge model. This was with a lower penalisation hyper parameter than the second Ridge model.

XGBoost regression did not perform as well as the Ridge model, but was nearly identical. It is often the case that simple models can outperform the more complicated models

Trying to get rid of the outliers in the training set actually led to an increase in the error of the testing set evaluation metric (RMSE). While the training set actually experienced a decline in RMSE, overfitting to the training data (which only had 134 observations with outliers removed) was harmful to the model.
